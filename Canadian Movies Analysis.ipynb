```python
# Canadian Movies Analysis
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline 


df= pd.read_csv(r'C:\Users\Admin\Desktop\AlexTheAnalyst\Python project\movies.csv\movies.csv')
canadian_movies = df[df['country'] == 'Canada']
canadian_movies
```

```python
for col in canadian_movies.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col,pct_missing))
```

```python
canadian_movies= canadian_movies.dropna(subset=['budget'])
```

```python
canadian_movies
```

```python
canadian_movies.boxplot(column=['gross'])
```

```python
canadian_movies.drop_duplicates()
```

```python
sns.regplot(x="gross", y="budget", data=canadian_movies)
```

```python
sns.regplot(x="score", y="gross", data=canadian_movies)
```

```python
canadian_movies.corr(method ='pearson')
```

```python
correlation_matrix = canadian_movies.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

canadian_movies.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = canadian_movies.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# top 15 compaies by gross revenue

CompanyGrossSum = canadian_movies.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
