```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\Admin\Desktop\Motor Viehcle Sales canada\Canada Vihecle - @nd Trial\20100001.csv")
data
```

```python
#Renaming
new_column_names = {
    "REF_DATE":"Year",
    "GEO" : "Province",
    "Vehicle type":"Type",
    "Origin of manufacture":"Origin",
    "VALUE" :"Sales_Thousands"
}
data.rename(columns=new_column_names, inplace=True)
data
```

```python

# Remove rows where "Units" is present in the "Sales" column
data.drop(data[data['Sales'].str.contains('Units')].index, inplace=True)

# Reset index after removing rows
data.reset_index(drop=True, inplace=True)

# Print the first few rows of the modified DataFrame
print(data.head())
```

```python
import pandas as pd
# Define the word and column to check
word_to_check = 'Units'
column_to_check = 'Sales'

# Filter out rows where the specified word exists in the specified column
data = data[~data[column_to_check].str.contains(word_to_check)]

# Reset index after removing rows
data.reset_index(drop=True, inplace=True)

# Print the first few rows of the modified DataFrame
print(data)
```

```python
import pandas as pd

# Keep only the Dollars values in the Sales Column
# Filter out rows where "Dollars" is not present in the "Sales" column
data = data[data['Sales'].str.contains('Dollars')]

# Reset index after filtering rows
data.reset_index(drop=True, inplace=True)

# Print the first few rows of the modified DataFrame
print(data.head())
```

```python

# Loop through each column and print its unique values
for column in data.columns:
    unique_values = data[column].unique()
    print(f"Unique values in '{column}': {unique_values}")
```

```python
# Define the words to check
words_to_check = ['North America', 'Total, overseas']

# Filter out rows where the 'Origin' column includes the specified words
data = data[data['Origin'].isin(words_to_check)]

# Reset index after filtering rows
data.reset_index(drop=True, inplace=True)

# Print the first few rows of the modified DataFrame
print(data.head())
```

```python
# Final Check on columns values and blanks
type_counts = data['Type'].value_counts()
print("Unique values in 'Type' column and their counts:")
print(type_counts)
```

```python
#Delete blanks, zeros or Nans
# Remove rows with blank, NaN, or zero values in the 'Type' column
data = data[data['Type'].notna() & (data['Type'] != '') & (data['Type'] != '0')]

# Reset index after removing rows
data.reset_index(drop=True, inplace=True)

# Print the first few rows of the modified DataFrame
print(data.head())
```

```python

# Convert 'Sales_Thousands' column to numeric format
data['Sales_Thousands'] = pd.to_numeric(data['Sales_Thousands'], errors='coerce')

# Group by 'Province' and 'Year', and sum the 'Sales_Thousands' values
aggregated_data = data.groupby(['Province', 'Year'])['Sales_Thousands'].sum().reset_index()

# Set the display format for float values
pd.options.display.float_format = '{:.2f}'.format

# Print the aggregated data
print(aggregated_data)
```

```python


# Convert 'Sales_Thousands' column to numeric format
data['Sales_Thousands'] = pd.to_numeric(data['Sales_Thousands'], errors='coerce')

# Filter out rows where 'Sales_Thousands' is NaN
data = data[data['Sales_Thousands'].notna()]

# Group by 'Province' and 'Year', and sum the 'Sales_Thousands' values
aggregated_data = data.groupby(['Province', 'Year'])['Sales_Thousands'].sum().reset_index()

# Set the display format for float values
pd.options.display.float_format = '{:.2f}'.format

# Print the aggregated data
print(aggregated_data)
```

```python
# Scatter plot of total sales vs. year with x-axis labels rotated by 60 degrees
plt.figure(figsize=(10, 6))
plt.scatter(aggregated_data['Year'], aggregated_data['Sales_Thousands'], alpha=0.6)
plt.title('Scatter Plot of Total Sales vs. Year')
plt.xlabel('Year')
plt.ylabel('Total Sales (Thousands)')
plt.xticks(rotation=60)  # Rotate x-axis labels by 60 degrees
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert 'Year' column to numeric format
aggregated_data['Year'] = pd.to_numeric(aggregated_data['Year'])

# Split data into two periods: before and after year 2000
before_2000 = aggregated_data[aggregated_data['Year'] < 2000]
after_2000 = aggregated_data[aggregated_data['Year'] >= 2000]

# Calculate the total sales for each province in each period
total_sales_before = before_2000.groupby('Province')['Sales_Thousands'].sum()
total_sales_after = after_2000.groupby('Province')['Sales_Thousands'].sum()

# Create a grouped bar plot for total sales before and after year 2000
plt.figure(figsize=(12, 8))
width = 0.35
x = range(len(total_sales_before))

plt.bar(x, total_sales_before, width, label='Before 2000')
plt.bar([i + width for i in x], total_sales_after, width, label='After 2000')
plt.title('Total Sales per Province: Before and After Year 2000')
plt.xlabel('Province')
plt.ylabel('Total Sales (Thousands)')
plt.xticks([i + width/2 for i in x], total_sales_before.index, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert 'Year' column to numeric format
aggregated_data['Year'] = pd.to_numeric(aggregated_data['Year'])

# Years to consider
years_to_plot = range(2019, 2023)

# Calculate the total sales for each province for each year
total_sales_per_year = {}
for year in years_to_plot:
    data_year = aggregated_data[aggregated_data['Year'] == year]
    total_sales_per_year[year] = data_year.groupby('Province')['Sales_Thousands'].sum()

# Create grouped bar plots for total sales for each year
num_years = len(years_to_plot)
width = 0.2
x = range(len(total_sales_per_year[years_to_plot[0]]))

plt.figure(figsize=(12, 8))

for idx, year in enumerate(years_to_plot):
    plt.bar([i + width * idx for i in x], total_sales_per_year[year], width, label=str(year))

plt.title('Total Sales per Province: 2019 to 2022')
plt.xlabel('Province')
plt.ylabel('Total Sales (Thousands)')
plt.xticks([i + width * (num_years - 1) / 2 for i in x], total_sales_per_year[years_to_plot[0]].index, rotation=45, ha='right')
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert 'Year' column to numeric format
aggregated_data['Year'] = pd.to_numeric(aggregated_data['Year'])

# Filter data for Atlantic provinces
atlantic_provinces = ['Newfoundland and Labrador', 'Prince Edward Island', 'Nova Scotia', 'New Brunswick']
data_atlantic = aggregated_data[aggregated_data['Province'].isin(atlantic_provinces)]

# Calculate the total sales for each year and province
total_sales_per_province_year = data_atlantic.groupby(['Province', 'Year'])['Sales_Thousands'].sum()

# Convert sales values to millions of dollars
total_sales_per_province_year = total_sales_per_province_year / 1000

plt.figure(figsize=(12, 6))

# Loop through each province and plot its sales over the years
for province in atlantic_provinces:
    province_data = total_sales_per_province_year.loc[province]
    plt.plot(province_data.index, province_data.values, marker='o', label=province)

plt.title('Total Sales for Atlantic Provinces: 1960 to 2022')
plt.xlabel('Year')
plt.ylabel('Total Sales (Millions of Dollars)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
